```python
def process_mag_data(myfile):
    print('Loading ',myfile)
    tdms_file = TdmsFile(myfile, memmap_dir='/data1/ryan')  #set buffer directory
    tddf = tdms_file.as_dataframe()  #LOADS FILE myfile AS DATAFRAME

    #ch_to_load1 = 'S_top' #acoustic channel 1
    #ch_to_load2 ='S_bot' #acoustic channel 2
    ch_to_load3 = 'IMAG' # magnet current channel, value needs to be multiplied by 1920 to 
                          #get current in Amperes
    #imag_sample = getchdata(tddf,ch_to_load3,1)[location-1000: location+4000]
    mag_current_au = getchdata(tddf,ch_to_load3,1)#load the whole thing?  
    mag_current_amps = 1920*mag_current_au

    #this has weird artificial zeros, how can we throw them out? or assign them more realistic values?
    new_points = []
    for i, point in enumerate(mag_current_amps):
        if point < 100:
            #set equal to its left neighbor (in the best way we can)
            new_point = new_points[i-1]
        else:
            new_point = point
        #overwrite outside if statement!    
        new_points.append(new_point)
    return np.asarray(new_points)
```

```python
myfile = '/data0/stephey/SORTED/8_18_2017_11_29_52_AM_A97_Q76_15163A.tdms'
q32_mag_current_nonzero = process_mag_data(myfile)
np.save('/home/ryanjia/datafiles/q06_mag_current_nonzero.npy',q32_mag_current_nonzero)
plt.plot(q32_mag_current_nonzero)
plt.show()
```
