```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
result = np.load('/data1/ryan/contexts/context_event_0.npy')
for i in range(1, 83):
    temp = np.load('/data1/ryan/contexts/context_event_' + str(i) + '.npy')
    result = np.concatenate((result, temp))
```

```python
embedded = TSNE(n_components=2, perplexity=10).fit_transform(result)
x = embedded[:, 0]
y = embedded[:, 1]
t = x

plt.figure()
plt.scatter(x, y, c=t)
plt.title('Perplexity = 10')
plt.show()
```
